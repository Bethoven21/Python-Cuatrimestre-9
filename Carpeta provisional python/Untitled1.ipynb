### Reto: Realizar un programa utilizando el ciclo anterior pero se debe solicitar al usuario el numero de veces que el usuario introducira valores y deben contar cuantos son pares y cuantos impares

```python
total_valores = int(input("Introduce la cantidad de valores que deseas ingresar: "))
pares = 0
impares = 0

for _ in range(total_valores):
    valor = int(input("Introduce un valor: "))
    if valor % 2 == 0:
        pares += 1
    else:
        impares += 1

print(f"Total de valores pares: {pares}")
print(f"Total de valores impares: {impares}")
```
